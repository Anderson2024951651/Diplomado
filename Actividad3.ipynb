Cargar la base de datos kmeans1.csv utilizando pandas.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.metrics import davies_bouldin_score
from yellowbrick.cluster import SilhouetteVisualizer
from yellowbrick.datasets import load_nfl
df = pd.read_csv('kmeans.1.csv', delimiter=',')
```

Por medio de matplotlib genere un gráfico de dispersión (scatter plot) de las variables A y B.

```python
df = (df - df.min()) / (df.max() - df.min()) #Escalamiento de Datos
plt.figure()
plt.scatter(df['A'], df['B'])
plt.show()
```

Pregunta: Aplique el algoritmo k-means (from sklearn.cluster import KMeans), observe y analice los outputs disponibles (centroides, clusters asignados y distancia dentro de las clases).
Ajuste k-means considerando 1,2,...10 clusters, guarde la distancia intra clases en un diccionario.
Utilizando matplotlib (o seaborn) genere un gráfico que presente la distancia intra clases para cada k utilizado. A través del criterio del codo defia el número de clusters óptimo.

```python
resultados={}
for k in range (1,11):
    k1=KMeans(n_clusters=k,max_iter=1000)
    k1.fit(df)
    resultados[k]=k1.inertia_
plt.figure()
plt.plot(list(resultados.keys()),list(resultados.values()),color='r')
plt.show()
print(k1.cluster_centers_)
print(k1.inertia_)
print(k1.labels_)

import Actividad3
Actividad3.evalua_clusters(df, [2,10])
```
